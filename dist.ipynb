```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.left'] = False
plt.rcParams["font.size"] = 15

def normal_dist(mu, sigma, xs):
  return 1 / np.sqrt(2 * np.pi) / (sigma) * np.exp(-1 / (2 * sigma ** 2) * (xs - mu) ** 2)

mu, sigma = 0, 0.2

xs = np.linspace(-1, 1, 300)
# ys = 1 / np.sqrt(2 * np.pi) * np.exp(-1 / 2 * xs ** 2)
ys = normal_dist(mu, sigma, xs)

# half of max
# hm = ys.max() / 2
hm = normal_dist(mu, sigma, 0.39199)
# indice of xs at which ys are greater than hm
fw = ys >= hm

# the left-most one
left = xs[fw][0]
# the right-most one
right = xs[fw][-1]

# horizontal and vertical delimiters
# plt.axhline(y=hm, linestyle='--', color='orange')
plt.axvline(x=left, linestyle='--', color='orange')
plt.axvline(x=right, linestyle='--', color='orange')

# shading
plt.fill_between(xs[fw], 0, ys[fw], color='g', alpha=0.3)
# plt.text(-0.1, 0.6, "$\\tau=95\%$")
# plt.text(-0.55, 0.3, "$-\delta$")
# plt.text(0.4, 0.3, "$\delta$")
plt.xticks([-0.4, 0, 0.4], labels=["$-\delta$", "0", "$\delta$"])
plt.yticks([])
plt.ylim([0,2])


plt.plot(xs, ys)
plt.spring

plt.savefig("norm_dist.pdf")
```
